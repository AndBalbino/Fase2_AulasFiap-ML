```python
from scipy.stats import shapiro
```

```python
stat, p = shapiro(df_studio)
print('Soma do barulho dos albuns de studio: ')
print('Estátistica de teste: {:.4f}, valor p: {:.10f}'.format(stat, p))

if p > 0.05:
  print('Não há evidência a hipotese de normalidade, distribuição normal')
else:
  print('A hipotese de normalidade é rejeitada, distribuição não normal')
```

```python
stat, p = shapiro(df_ao_vivo)
print('Soma do barulho dos albuns ao vivo')
print('Estátistica de teste: {:.4f}, valor p: {:.10f}'.format(stat, p))

if p > 0.05:
  print('Não há evidência a hipotese de normalidade, distribuição normal')
else:
  print('A hipotese de normalidade é rejeitada, distribuição não normal')
```

```python
from scipy.stats import mannwhitneyu
```

```python
stat, p = mannwhitneyu(df_studio.sample(len(df_studio)), df_ao_vivo.sample(len(df_ao_vivo)), alternative='less')

print('Estatistica de teste U: ', stat)
print('Valor p: ', p)

alpha = 0.05

if p < alpha:
  print('Diferença estatísticamente significante')
else:
  print('Não há diferença estatísticamente significativa')
```

### Modelo de ML Regressão linear

```python
from sklearn.linear_model import LinearRegression
```

```python
# Criando um obj de regressão
lr = LinearRegression()
```

```python
# X contem as variaveis preditoras ou independentes
X = df_ambiental[['Area', 'Suites', 'IA', 'Vista', 'Andar', "AV100m", 'DistBM', 'Semruido']]

# Y variavel target ou dependente

y = df_ambiental[['Valor']]
```

```python
from sklearn.model_selection import train_test_split

# Separando os dados em Treino e Teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# Treinando o Modelo
lr.fit(X_train, y_train)
```

```python
# Calculando o valor predito da variavel resposta na amostra teste
y_pred = lr.predict(X_test)
```

```python
# Primeiro, vamos olhar o Intercepto e os Coeficientes de Regressão.
# Representa o valor esperado da variável dependente quando todas as variáveis independentes são iguais
# Em termos gráficos, o intercepto é o ponto onde a linha de regressão cruza p eixo vertical (eixo y)

print("Interceptor: ", lr.intercept_)
```

```python
coefficients = pd.concat([pd.DataFrame(X.columns), pd.DataFrame(np.transpose(lr.coef_))], axis=1)
coefficients
```

```python
fig = plt.figure(figsize=(8,6), dpi=80)
plt.rcParams.update({"font.size": 14})
ax = sns.regplot(x = y_test, y = y_pred)
ax.set(xlabel= 'x real', ylabel= 'y predito')
ax = plt.plot(y_test, y_test, '--r')

plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Avaliando o modelo
MSE = mean_squared_error(y_test, y_pred)
MAE = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MAE: ", MAE) # Mean absolute error (MAE) é a média do valor absoluto dos erros.
print("MSE: ", MSE) # Erro quadrático médio (MSE) é a média dos erros quadráticos
print("R2: ", r2) # R-quardrado # Próximo de 1 mas nunca 1
```

```python
# Convertento as colunas que contém caracteres para números

df_airlines['AirportFrom']=LabelEncoder().fit_transform(df_airlines['AirportFrom'])
df_airlines['AirportTo']=LabelEncoder().fit_transform(df_airlines['AirportTo'])
df_airlines['Airline']=LabelEncoder().fit_transform(df_airlines['Airline'])
```

```python
df_airlines.head()
```

### Separando a base de dados

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_airlines[['Flight', 'Airline', 'AirportFrom', 'AirportTo', 'DayOfWeek', 'Time', 'Length']]
y = df_airlines['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=7)
```

### Criando o modelo de Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
# Predizendo valores
y_pred = gnb.predict(x_test)
```

### Validando o modelo

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", round(accuracy,2))
```

### Equilibrando a base de dados

```python
from sklearn.utils import resample
```

```python
# Separando as classes majoritárias e minoritárias
df_majority = df_airlines[df_airlines.Class==0]
df_minority = df_airlines[df_airlines.Class==1]

# Dupplicar os dados e criar uma nova representação quase que igualitária
```

```python
# Upsampling da classe minoritária
df_minority_upsampled = resample(df_minority, replace=True, n_samples=len(df_majority), random_state=123)
```

```python
# Jutando os dois DataFrames
df_equilibrado = pd.concat([df_majority, df_minority_upsampled])
```

```python
sns.countplot(x='Class',  data=df_equilibrado)
plt.show()
```

### Testando o algoritmo com a base equilibrada

```python
x_equilibrado = df_equilibrado[['Flight', 'DayOfWeek', 'Time', 'Length']]
y_equilibrado = df_equilibrado['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_equilibrado,y_equilibrado, test_size=0.2, random_state=7, stratify=y_equilibrado)
```

```python
# treinando o algoritmo
gnb.fit(x_train, y_train)
```

```python
# Predizendo valores
y_pred = gnb.predict(x_test)
```
